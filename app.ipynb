Distribución Geométrica

p(r;p)=p(1−p)r−1
 
En dónde r≥1 y el parámetro p es un número real. La Distribución Geométrica expresa la probabilidad de tener que esperar exactamente r pruebas hasta encontrar el primer éxito si la probabilidad de éxito en una sola prueba es p. Por ejemplo, en un proceso de selección, podría definir el número de entrevistas que deberíamos realizar antes de encontrar al primer candidato aceptable.
Por ejemplo, si estás estudiando el tiempo entre tornados significativos en una región determinada y quieres saber cuánto tiempo pasará hasta el próximo tornado importante, puedes utilizar una distribución geométrica.

PRIMERA PARTE
-import numpy as np
-from scipy import stats
-import matplotlib.pyplot as plt
En estas líneas, se importan las bibliotecas necesarias: numpy para operaciones numéricas, scipy.stats para la distribución geométrica y matplotlib.pyplot para generar gráficos.

-p = 0.3  # Parámetro de forma
-geometrica = stats.geom(p)  # Distribución
Aquí se define el parámetro de forma p para la distribución geométrica y se crea un objeto geometrica que representa la distribución geométrica utilizando el parámetro p.

-x = np.arange(geometrica.ppf(0.01), geometrica.ppf(0.99))
Se genera un arreglo x que contiene los valores en el eje x del gráfico. Estos valores se calculan utilizando el método ppf (Percent Point Function) de la distribución geométrica. En este caso, se generan valores entre el percentil 0.01 y el percentil 0.99 de la distribución.

-fmp = geometrica.pmf(x)  # Función de Masa de Probabilidad
Se calcula la función de masa de probabilidad (PMF) para los valores en x utilizando el método pmf de la distribución geométrica. La PMF representa la probabilidad de obtener cada valor en x en la distribución.

-plt.plot(x, fmp, '--')
Se traza una línea que muestra la función de masa de probabilidad en el gráfico. Los valores de x están en el eje x y los valores de fmp en el eje y.

-plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
Se trazan líneas verticales desde el eje x hasta los puntos en la función de masa de probabilidad. Estas líneas resaltan la probabilidad asociada con cada valor en x.

-plt.title('Distribución Geométrica')
-plt.ylabel('probabilidad')
-plt.xlabel('valores')
Estas líneas establecen los títulos de los ejes y el título del gráfico.

-plt.show()
Finalmente, se muestra el gráfico resultante.


SEGUNDA PARTE
-np.random.seed(0)
Esta línea establece la semilla aleatoria para reproducibilidad. Al fijar una semilla, los números aleatorios generados serán los mismos cada vez que se ejecute el código.

-aleatorios = geometrica.rvs(100)  # genera aleatorios
Aquí se generan 100 números aleatorios utilizando el método rvs (Random Variates) de la distribución geométrica (geometrica). Los números aleatorios generados siguen la distribución geométrica especificada anteriormente.

-cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
Esta línea genera y muestra un histograma de los números aleatorios generados (aleatorios). El histograma divide los valores en 20 "cajas" (bins) y cuenta cuántos valores caen en cada caja. La función hist devuelve tres valores: cuenta (un arreglo con las frecuencias de cada caja), cajas (los bordes de las cajas) y ignorar (un objeto de parche que se ignora en este caso).

-plt.ylabel('frequencia')
-plt.xlabel('valores')
-plt.title('Histograma Geométrica')
-plt.show()
Estas líneas establecen los títulos de los ejes (plt.ylabel y plt.xlabel) y el título del gráfico (plt.title). Finalmente, plt.show() muestra el histograma resultante.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


# Graficando Geométrica
p =  0.4 # parametro de forma 
geometrica = stats.geom(p) # Distribución
x = np.arange(geometrica.ppf(0.001),
              geometrica.ppf(0.99))
y = geometrica.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, y)
plt.vlines(x, 0, y, colors='darkblue', lw=1, alpha=0.5)
plt.title('Distribución Geométrica')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
# histograma
np.random.seed(0)
aleatorios = geometrica.rvs(1000)  # genera aleatorios
plt.hist(aleatorios, 20, color='darkblue')
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Distribución Geométrica')
plt.show()
```
